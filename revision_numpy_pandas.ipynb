```python
import numpy as np
import pandas as pd
```

```python
#creation of numpy array with python object
a=np.array([[1,2,3],[4,5]],dtype="object")
print(a)
print('*********************')
#creation of numpy array with zeros
a=np.zeros((3,4))
print(a)
print('*********************')
#creation of numpy array with ones 
a=np.ones((3,4))
print(a)
print('*********************')
#creation of numpy array with arange
a=np.arange(1,10)
print(a)
print('*********************')
#creation of numpy array with identity matrix
a=np.eye(3,3)
print(a)
```

```python
#Array attributes
#dtype is an attribute in numpy which is used to get type of an array object
#shape is an attribute in numpy which is used to get shape of array which will have information about how many rows and columns are there (row,column)
#size is an attribute in numpy which will give total number of elements present in an array
#ndim is an attribute in numpy which will give dimension of an numpy array
a=np.array([[1,2,3],[4,5,6]])
print(a.dtype)
print('*********************')
print(a.shape)
print('*********************')
print(a.size)
print('*********************')
print(a.ndim)
```

```python
#reshape function is used to change the shape of the array but the number of element should be same in new shape also
#flatten function is used to change the shape of the array to one dimension directly
#stack function is used to join the two array into single araay but major difference is it will join new array as new cell that is dimension will change,increase
#concatenate function will add another array rows in to an array that is dimension will not change another array rows will be concatenated to existed one from end
a=np.array([[1,2,3],[4,5,6]])
print(a.reshape((3,2)))
print('*********************')
print(a.flatten())
print('*********************')
print(np.stack((a,a)))
print('*********************')
print(np.concatenate((a,a)))
```

```python
#array indexing and slicing 
a=np.array([[1,2,3],[4,5,6]])
# give the row and column index as such : [row,column]
print(a[1,1])
print('*********************')
# give range of rows and column indices as such :[starting row index: ending row index, starting column index: ending column index]
print(a[:,1:])
print('*********************')
# give boolean expression within square brackets to fetch some elements from array
print(a[a%2==0])
```

```python
#mathematical operations and statistical operation
a=np.array([[1,2,3],[4,5,6]])
# to add certain value to each element of array
print(a+2)
print('*********************')

# use np.sum() to get sum of all elements of array
print(np.sum(a))
print('*********************')

# use np.min() to get minimum element of the array
print(np.min(a))
print('*********************')

# use np.max() to get maximum element of the array
print(np.max(a))
print('*********************')

# use np.argmin() to get index of minimum element of the array
print(np.argmin(a))
print('*********************')

# use np.argmax() to get index of maximum element of the array
print(np.argmax(a))
print('*********************')

# use np.sqrt() to get square-root of all elements of the array
print(np.sqrt(a))
print('*********************')

# use np.exp() to get exponential of all elements of the array
print(np.exp(a))
print('*********************')

# use np.log() to get logarithm of all elements of the array
print(np.log(a))
print('*********************')

# use np.sin() to get sine of all elements of the array
print(np.sin(a))
print('*********************')

# use np.cos() to get cos of all elements of the array
print(np.cos(a))
print('*********************')

# use np.tan() to get tan of all elements of the array
print(np.tan(a))
print('*********************')

# use np.mean() to get mean of all elements of the array
print(np.mean(a))
print('*********************')

# use np.median() to get median of all elements of the array
print(np.median(a))
print('*********************')

# use np.var() to get variance of all elements of the array
print(np.var(a))
print('*********************')

# use np.cov() to get covariance of all elements of the array
print(np.cov(a))
print('*********************')

# use np.std() to get standard deviation of all elements of the array
print(np.std(a))
print('*********************')

# use np.dot() to get dot product of two arrays
b = np.array([[1,2],[3,4]])
print(np.dot(b,b))
print('*********************')

# use np.matmul() to get matrix multiplication of two arrays
print(np.matmul(b,b))
print('*********************')

# use np.linalg.norm() to get normalised distance of the array
print(np.linalg.norm(a))
print('*********************')

# use np.linalg.inv() to get inverse of the array
print(np.linalg.inv(b))
print('*********************')
```

```python
#array manipulation
a=np.array([[1,2,3],[4,5,6]])
#np.transpose() function will change the shape of array such that rows and columns are interchanged.
print(np.transpose(a))
print('*********************')

#np.hstack() will stack the arrays horizontally
print(np.hstack((a,a)))
print('*********************')

#np.vstack() will stack the arrays vertically
print(np.vstack((a,a)))
print('*********************')

#np.split() will split the array at specified index
print(np.split(a,[1]))
print('*********************')

#np.delete() will delete the elements at specified index
print(np.delete(a,[1,2]))
print('*********************')

#np.insert() will insert the elements at specified index
print(np.insert(a,2,10))
print('*********************')

#np.append() will append elements at the end
print(np.append(a,10))
```

```python
#random generation
#np.random.rand() will generate a random numpy array of specified shape
print(np.random.rand(2,3))
print('*********************')

#np.random.randint() will generate a random numpy array of specified shape with values between the two numbers
print(np.random.randint(2,10,size=(2,2)))
print('*********************')

#np.random.choice() will select some number of arrays from already existing 1-dimensional array
print(np.random.choice(a.flatten(),size=4))
```

```python
#sorting
a=np.array([[3,2,1],[6,4,5]])
#use np.sort() to sort the array in ascending order
print(np.sort(a))
print('*********************')

#np.argsort() will give the indexes of unsorted array
print(np.argsort(a))
```

```python
#selection and indexing
df=pd.read_csv("BangaloreTopCompanies.csv")
#selecting single column
print(df["CompanyName"])
print('*********************')

#selecting multiple columns
print(df[["CompanyName","Rating","Location"]])
print('*********************')

#selecting rows using index
print(df[10:13])
print('*********************')

#accesing specific element by using label using loc[]
print(df.loc[10,"CompanyName"])
print('*********************')

print(df.loc[10,["CompanyName","Location"]])
print('*********************')

#accesing specific element by using index using iloc[]
print(df.iloc[490,1])
print('*********************')

print(df.iloc[490:499,1:4])


```

```python
#filtering 
df=pd.read_csv("BangaloreTopCompanies.csv")
#filtering based on single condition
print(df[df["Location"]=='Bengaluru'])
print('*********************')

#filtering based on multiple condition with and ,or operators
print(df[(df["Rating"]==4.1)|(df["Location"]=='Bengaluru')])
print('*********************')

print(df[(df["Industry"]=="Business Consulting")&(df["Location"]=='Bengaluru')])
print('*********************')

print(df.query("Rating==4.1 or Location=='Bengaluru'"))
print('*********************')

print(df.query("Rating==4.1 and Location=='Bengaluru'"))
print('*********************')


#filtering nan rows
print(df[df["CompanyName"].isna()])
print('*********************')

print(df[df["CompanyName"].notna()])
print('*********************')

#fetching rows with more than one nan in present in data frame
print(df[df.isna().sum(axis=1)>=1])
```

```python
#data cleaning
df=pd.read_csv("BangaloreTopCompanies.csv")
#dropna() is function in pandas which will remove the rows from data frame if one nan is present
print(df.dropna())
print('*********************')

print(df.dropna(subset=["Description"]))
print('*********************')

print(df.dropna(thresh=8))
print('*********************')

#drop() is function in pandas which will remove rows and columns with specified axis with column label for column and row index for row
print(df.drop("Unnamed: 0",axis=1))
print('*********************')

print(df.drop(1,axis=0))
print('*********************')

#replace is function in pandas which is used to change the old value into new value with respect to entire data frame and also specified feature of a data frame
print(df.replace("Bengaluru","Bangalore"))
print('*********************')

df["Location"]=df["Location"].replace("Bangalore","Bengaluru")
print(df)
print('*********************')

#rename is function pandas which is used to change names of columns  
print(df.rename(columns={"CompanyName":"company_name"}))
print('*********************')

#sort_values is function in pandas which is used to sort the data frame by specified columns
print(df.sort_values(by=["CompanyName","Rating"]))
print('*********************')

print(df.sort_index())
```

```python
#changing type of the column data
df=pd.read_csv("BangaloreTopCompanies.csv")
df['Rating']=df['Rating'].astype("str")
print(df['Rating'].dtype)
```

```python
#groupby() is a function in pandas which is used to group the categorical features and perform aggregate operations as needed with help of agg() function
#after doing groupby() to the df[<column>] it will return a groupby object. Later we can apply agg() function to perform aggregate operations with respect to another column.
a={"id":[1,2,3,4,5,6],"Name":["shashank","rashmi","prathap","sushmitha","savitha","kavitha"],"work":["not working","working","not working","working","working","not working"],"gender":["M","F","M","F","F","F"]}
df1 = pd.DataFrame(a)
b=df1.groupby("gender")
print(b.agg({"id":"count"}))
print('*********************')


#pivot_table() is a function in pandas used to perform the same operation as groupby in a simple way.
print(df1.pivot_table(index='gender',values='id',aggfunc='count'))
```

```python
# pd.concat() is a function in pandas used to concatenate two dataframes into a single dataframe with reference to axis i.e., axis=0 or 1
b=[{"id":1,"name":"shashank","age":24},{"id":2,"name":"divya","age":25},{"id":3,"name":"rashmi","age":23},{"id":4,"name":"kavya","age":21}]
df2 = pd.DataFrame(b)
print(pd.concat([df1,df2],axis=1))
print('*********************')


#pd.merge() is a function in pandas used to merge two dataframes into a single dataframe with related columns. Merge will work same as SQL joins.
print(pd.merge(df1,df2,right_on='name',left_on='Name'))
print('*********************')

print(pd.merge(df1,df2,right_on='name',left_on='Name',how='outer'))
print('*********************')

print(pd.merge(df1,df2,right_on='name',left_on='Name',how='right'))
print('*********************')

print(pd.merge(df1,df2,right_on='name',left_on='Name',how='left'))
```

```python
#apply() is a function in pandas which is used to apply the custom function to the dataframe column and also to the entire dataframe

def change_gender_value(gender):
    if gender=='M':
        return 'Male'
    elif gender=='F':
        return 'Female'
    else:
        return 'Other'

df1['gender']=df1['gender'].apply(change_gender_value)
print(df1)
```
